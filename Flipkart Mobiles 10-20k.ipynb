```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:/Users/navne/Desktop/Flipart_Mobiles-10-20k.csv")
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.drop_duplicates(['rating&review_c']).head()
```

```python
df.isnull().sum()
```

# Cleaning Name 

```python
df['color'] = df['name'].apply(lambda x:x.split("(")[1].split(",")[0]) #got color of the product
```

```python
df['rom'] = df['name'].apply(lambda x:x.split(",")[1].strip().split(" ")[0]) #got rom of the product
```

```python
df['name1'] = df['name'].apply(lambda x:x.split("(")[0].strip())
```

```python
df.drop(['name'],axis=1,inplace=True) #name column is cleaned
```

# Cleaning Price 

```python
df['price'] = df['price'].apply(lambda x:x.replace('₹',''))
```

```python
df.head()
```

# Cleaning Rating & Review Count

```python
df['rating_count'] = df['rating&review_c'].apply(lambda x:x.split('&')[0].split(" ")[0])
```

```python
df['review_count'] = df['rating&review_c'].apply(lambda x:x.split('&')[1].strip().split(" ")[0])
```

```python
df.drop(['rating&review_c'],axis=1,inplace=True)
```

# Cleaning Description

```python
df['description']
```

```python
df['ram'] = df['description'].apply(lambda x:x.split("|")[0].strip(" ")[0])
```

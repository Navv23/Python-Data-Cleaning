# 2. Checking Nulls and Duplicates

```python
df.isnull().sum()
```

```python
df['rating'].unique()
```

```python
def handlerating(x):
    if x=='NEW' or x=='-':
        return np.nan
    else:
        x = str(x).split('/')
        x = x[0]
        return float(x)
```

```python
df['rating'] = df['rating'].apply(handlerating)
```

```python
df
```

```python
df1 = df.dropna(how = 'any')
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
df['name'] = df['name'].drop_duplicates()
```

```python
df['approx_cost_2ppl'].unique()
```

```python
def handlecost(x):
    x = str(x)
    if ',' in x:
        x = x.replace(',','')
        return float(x)
    else:
        return float(x)
```

```python
df['approx_cost_2ppl'] = df['approx_cost_2ppl'].apply(handlecost)
```

# 3. Examning the Outliers of Each Column

```python
df.head()
```

```python
rest_types = df['rest_type'].value_counts()
```

```python
rest_types
```

```python
rest_lessthan_1000 = rest_types[rest_types < 1000]
```

```python
def handlerest_type(x):
    if x in rest_lessthan_1000:
        return 'Other'
    else:
        return x
```

```python
df['rest_type'] = df['rest_type'].apply(handlerest_type)
```

```python
df['rest_type'].value_counts()
```

```python
loc = df['location'].value_counts()
```

```python
loclessthan_300 = loc[loc<300]
```

```python
def handlelocation(x):
    if x in loclessthan_300:
        return 'Other'
    else:
        return x
```

```python
df['location'] = df['location'].apply(handlelocation)
```

```python
df['location'].value_counts()
```
